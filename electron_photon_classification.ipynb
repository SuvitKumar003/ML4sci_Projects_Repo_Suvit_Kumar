```python
import os
import numpy as np
import h5py
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import transforms
from sklearn.metrics import classification_report, confusion_matrix
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
class Config:
    # Data
    ELECTRON_PATH = "./dataset/SingleElectronPt50_IMGCROPS_n249k_RHv1.hdf5"
    PHOTON_PATH = "./dataset/SinglePhotonPt50_IMGCROPS_n249k_RHv1.hdf5"
    BATCH_SIZE = 128
    TRAIN_RATIO = 0.8
    
    # Model
    INPUT_CHANNELS = 2
    IMAGE_SIZE = 32
    NUM_CLASSES = 2
    
    # Training
    EPOCHS = 30
    LEARNING_RATE = 0.001
    WEIGHT_DECAY = 1e-4
    PATIENCE = 5  # For early stopping
    
    # System
    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    SEED = 42
```

```python
torch.manual_seed(Config.SEED)
np.random.seed(Config.SEED)
```

```python
class ParticleDataset(Dataset):
    def __init__(self, electron_path, photon_path, transform=None):
        self.transform = transform
        
        # Open HDF5 files once and keep references
        self.electron_file = h5py.File(electron_path, 'r')
        self.photon_file = h5py.File(photon_path, 'r')
        
        # Store dataset references
        self.electron_data = self.electron_file['X']
        self.photon_data = self.photon_file['X']
        
        # Store dataset lengths
        self.electron_len = len(self.electron_data)
        self.photon_len = len(self.photon_data)
        self.total_len = self.electron_len + self.photon_len
```

```python

def __len__(self):
    return self.total_len
```

```python
def __getitem__(self, idx):
        # Read from disk on demand
        if idx < self.electron_len:
            data = self.electron_data[idx]  # Shape: (32, 32, 2)
            label = 0
        else:
            data = self.photon_data[idx - self.electron_len]
            label = 1

        if self.transform:
            data = self.transform(data)
            
        return data, torch.tensor(label).long()
```

```python
 def __del__(self):
        # Cleanup when dataset is garbage collected
        self.electron_file.close()
        self.photon_file.close()
```

```python
def create_data_loaders():
    # Calculate normalization constants
    def calculate_mean_std(dataset):
        # Use partial dataset to save memory
        subset_size = min(10000, len(dataset))
        subset = torch.stack([dataset[i][0] for i in range(subset_size)])
        return subset.mean(dim=(0,2,3)), subset.std(dim=(0,2,3))
    
    # Initialize dataset without transforms
    raw_dataset = ParticleDataset(
        Config.ELECTRON_PATH,
        Config.PHOTON_PATH,
        transform=transforms.ToTensor()
    )
    
    # Calculate dataset statistics
    mean, std = calculate_mean_std(raw_dataset)
    print(f"Calculated mean: {mean.tolist()}, std: {std.tolist()}")

    # Full transform pipeline
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomVerticalFlip(p=0.5),
        transforms.Normalize(mean=mean, std=std)
    ])
```

```python
# 1. Create temporary dataset to calculate stats
temp_dataset = ParticleDataset(
    Config.ELECTRON_PATH,
    Config.PHOTON_PATH,
    transform=transforms.ToTensor()
)

# 2. Calculate mean/std
loader = DataLoader(temp_dataset, batch_size=256)
mean = 0.0
std = 0.0
for images, _ in loader:
    mean += images.mean((2, 3)).sum(0)
    std += images.std((2, 3)).sum(0)
mean /= len(loader.dataset)
std /= len(loader.dataset)

# 3. Define final transform
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
])

# 4. Create final dataset
dataset = ParticleDataset(
    Config.ELECTRON_PATH,
    Config.PHOTON_PATH,
    transform=transform
)
```
